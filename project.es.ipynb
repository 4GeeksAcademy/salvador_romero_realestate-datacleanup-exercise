#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
# TODO
def normalizar(serie):
    new_serie = []
    for s in serie:
        new_serie.append(float((s-min(serie))/(max(serie)-min(serie))))
    return (new_serie)

ds_alc = ds_sur.loc[ds_sur["level5"]=="Alcorcón"]["price"]
ds_fuen = ds_sur.loc[ds_sur["level5"]=="Fuenlabrada"]["price"]
ds_get = ds_sur.loc[ds_sur["level5"]=="Getafe"]["price"]
ds_leg = ds_sur.loc[ds_sur["level5"]=="Leganés"]["price"]

fig,axs = plt.subplots(2,2)
axs[0,0].hist(normalizar(ds_alc))
axs[0,0].set_title("Alcorcon")
axs[0,1].hist(normalizar(ds_fuen))
axs[0,1].set_title("Fuenlabrada")
axs[1,0].hist(normalizar(ds_get))
axs[1,0].set_title("Getafe")
axs[1,1].hist(normalizar(ds_leg))
axs[1,1].set_title("Leganes")
for ax in axs.flat:
    ax.label_outer()
```

Una conclusión de estas gráficas es que en Getafe hay una variedad en los precios de las viviendas. En Alcorcón y Fuenlabrada los precios están más parejos.

#### Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón sur de Madrid por color de cada población (debes usar el subconjunto obtenido del Ejercicio 14) (★★★★)

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, basemaps,Marker

# Mapa centrado en (60 grados latitud y -2.2 grados longitud)
# Latitud, longitud
map = Map(center = (60, -2.2), zoom = 2, min_zoom = 1, max_zoom = 20)
map
```

```python
## Aquí: traza la coordenadas de los estados
from ipywidgets import HTML
## PON TU CÓDIGO AQUÍ:
# Getafe
lat_get = ds_sur.loc[ds_sur["level5"]=="Getafe"]["latitude"]
lon_get = ds_sur.loc[ds_sur["level5"]=="Getafe"]["longitude"]
# Alcorcon
lat_al = ds_sur.loc[ds_sur["level5"]=="Alcorcon"]["latitude"]
lon_al = ds_sur.loc[ds_sur["level5"]=="Alcorcon"]["longitude"]
# Fuenlabrada
lat_fue = ds_sur.loc[ds_sur["level5"]=="Fuenlabrada"]["latitude"]
lon_fue = ds_sur.loc[ds_sur["level5"]=="Fuenlabrada"]["longitude"]
# Leganes
lat_leg = ds_sur.loc[ds_sur["level5"]=="Leganes"]["latitude"]
lon_leg = ds_sur.loc[ds_sur["level5"]=="Leganes"]["longitude"]

for lat,lon in zip(lat_get,lon_get):
    marker_get = Marker(location=(lat, lon), draggable=False)
    marker_get.popup = HTML(value="Getafe") 
    map.add_layer(marker_get)

for lat,lon in zip(lat_fue,lon_fue):
    marker_fue = Marker(location=(lat, lon), draggable=False)
    marker_fue.popup = HTML(value="Fuenlabrada") 
    map.add_layer(marker_fue)

for lat,lon in zip(lat_al,lon_al):
    marker_al = Marker(location=(lat, lon), draggable=False)
    marker_al.popup = HTML(value="Alcorcon") 
    map.add_layer(marker_al)

for lat,lon in zip(lat_leg,lon_leg):
    marker_leg = Marker(location=(lat, lon), draggable=False)
    marker_leg.popup = HTML(value="Leganés") 
    map.add_layer(marker_leg)
map
```
